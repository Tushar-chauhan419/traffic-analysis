```python
# ðŸ“Š Website Traffic Analysis

# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load Data
df = pd.read_csv('/content/traffic.csv')  # Adjust if running on local
df.head()
```

```python
# Step 3: Data Cleaning
df.info()
df['date'] = pd.to_datetime(df['date'])  # Convert date column

# Check for missing values
print("Missing values:\n", df.isnull().sum())

# Drop duplicates if any
df = df.drop_duplicates()
```

```python
# Step 4: Traffic Over Time (Daily)
daily_traffic = df.groupby('date').size()
plt.figure(figsize=(14, 6))
daily_traffic.plot(kind='line', color='blue')
plt.title('ðŸ“… Website Traffic Over Time (Daily)')
plt.xlabel('Date')
plt.ylabel('Number of Clicks')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Step 5: Top 10 Countries by Clicks
top_countries = df['country'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_countries.plot(kind='bar', color='skyblue')
plt.title('ðŸŒ Top 10 Countries by Website Clicks')
plt.xlabel('Country')
plt.ylabel('Clicks')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Step 6: Top 10 Cities by Clicks
top_cities = df['city'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_cities.plot(kind='barh', color='orange')
plt.title('ðŸ™ï¸ Top 10 Cities by Clicks')
plt.xlabel('Clicks')
plt.ylabel('City')
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```

```python
# Step 7: Most Popular Tracks
top_tracks = df['track'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_tracks.plot(kind='bar', color='green')
plt.title('ðŸŽ¶ Top 10 Most Clicked Tracks')
plt.xlabel('Track')
plt.ylabel('Clicks')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Step 8: Most Popular Artists
top_artists = df['artist'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_artists.plot(kind='barh', color='purple')
plt.title('ðŸŽ¤ Top 10 Most Popular Artists by Clicks')
plt.xlabel('Clicks')
plt.ylabel('Artist')
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```

```python
# Step 9: Correlation (if applicable)
numeric_cols = df.select_dtypes(include='number')

if not numeric_cols.empty:
    plt.figure(figsize=(8, 5))
    sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm')
    plt.title("ðŸ” Correlation Heatmap")
    plt.show()
else:
    print("No numeric columns to analyze correlation.")
```
